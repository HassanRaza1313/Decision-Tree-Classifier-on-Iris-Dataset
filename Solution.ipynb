# Import Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OrdinalEncoder
```

# Load Dataset

```python
df = pd.read_csv("./Iris.csv")
df.head()
```

# Encode Species as Numerical Value

```python
encoder = OrdinalEncoder()
y = encoder.fit_transform(df[["Species"]])
y.reshape(-1,)
print(y.shape)
df.tail()
```

# Seperate Input and Output Values

```python
x = df.drop('Id',axis=1)
x = x.drop('Species',axis=1)
print(x)
```

# Extracting Feature and Target Names

```python
features = df.columns.to_list()[1:-1]
print(features)

target = df["Species"].unique()
print(target)
```

# Load Model

```python
model = DecisionTreeClassifier()
model = model.fit(x,y)
```

# Visualize and Save Decision Tree

```python
figure = plt.figure(figsize=(25,20))
tr = tree.plot_tree(model,feature_names=features,class_names=target,filled=True)
figure.savefig("decision_tree.jpeg")
```
